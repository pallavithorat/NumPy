Section 4 — Vectorization vs Loops

Task 4.1: Loop vs Vectorized

```python
# TODO:
# Create large array X of size 1,000,000
rng = np.random.default_rng(2) # random number generator
X = rng.standard_normal(1_000_000) # array of 1 million random numbers
```

```python
# TODO:
# Normalize using Python loop

# HINT: 
# - Time it
mu = X.mean() # mean
sigma = X.std() # standard deviation = how spread out the numbers are/ aaverage distance from mean
t0 = time.time() # start time
out_loop = np.empty_like(X) # output array

for i in range(X.shape[0]):
    out_loop[i] = (X[i] - mu) / sigma
time_taken = time.time() - t0
print(round(time_taken, 4))


```

```python
# TODO:
# Normalize using vectorization
t1 = time.time()
out_vec = (X - mu) / sigma
vec_time = time.time() - t1
round(vec_time, 4)
print(round(time_taken / max(vec_time, 1e-12), 1))
print('allclose:', np.allclose(out_loop, out_vec))

```

Why is vectorization faster?


Task 4.2: Pairwise Distance (FAANG Classic)

```python
# TODO:
# Compute pairwise Euclidean distance matrix without loops

# HINT:
# - Use (x - y)^2 expansion
# - Broadcasting is key

def pairwise_distance(X):

    sq = np.sum(X * X, axis=1, keepdims=True)  # shape (n, 1)
    dist = sq + sq.T - 2 *(X @ X.T)
    dist = np.maximum(dist, 0.0)  
    return np.sqrt(dist)

rng = np.random.default_rng(3)
Xsmall = rng.standard_normal((200, 10))
D = pairwise_distance(Xsmall)
print(D.shape)  # should be (200, 200)
print('diag ~ 0 :' , np.allclose(np.diag(D), 0, atol=1e-7))  # diagonal should be ~0
print('symmetry :', np.allclose(D, D.T))  # should be symmetric
    
```

Section 5 — Numerical Stability =  It refers to how sensitive a algorithm is in response to small errors.

Task 5.1: Softmax

```python
# TODO:
# Implement naive softmax

def softmax_naive(X):
    ex = np.exp(X)

    return ex/ex.sum(axis=0, keepdims=True)

Xbig = np.array([[1000.0, 1001.0, 1002.0]])

try:
    softmax_naive(Xbig)
except FloatingPointError as e:
    print('Numerical issue in softmax_naive:', e)
```

```python
# TODO:
# Fix numerical instability

# HINT:
# - Subtract max per row

def softmax_stable(X):
    Z = X - X.max(axis=1, keepdims=True)
    print('Z:', Z)
    ex = np.exp(Z)
    return ex / ex.sum(axis=1, keepdims=True)

print('stable softmax:', softmax_stable(Xbig))
print('rowsum :', softmax_stable(Xbig).sum(axis=1))
```

Why does subtracting max work?

```python
# TODO:
# Try valid and invalid matrix multiplications
# Matrix Multiplication : It's only possible when columns of first matrix = rows of second matrix
A = np.random.default_rng(3).standard_normal((3, 4))

B = np.random.default_rng(4).standard_normal((4, 2))

mat_mul = A @ B  # valid multiplication
print( mat_mul.shape)


try:
    _ = A @ A  # invalid multiplication
except ValueError as e:
    print('Matrix multiplication error:', e)


# @, np.dot, and np.matmul are equivalent for 2D arrays, difference is @ is easier to read and write
# matmul is older version
print('A @ B ==. dot.product', np.allclose(np.dot(A, B), mat_mul)) # true
print('matmul == A @ B', np.allclose(np.matmul(A, B), mat_mul)) # true
```

Task 7.1: In-Place Operations

```python
# TODO:
# Compare in-place vs out-of-place operations

X = np.random.default_rng(7).standard_normal(2_000_000)
Y = X.copy() # when i create a copy they both (X & Y) occupy different memory locations
t0 = time.time()
Z = X + 1.0  # out-of-place = creates a new array
t_out = time.time() - t0

t1 = time.time()
Y += 1.0  # in-place 
t_in = time.time() - t1

print('out-of-place time:', round(t_out, 4))
print('in-place time:', round(t_in, 4))
# In-place is generally faster and uses less memory since it modifies the existing array rather than creating a new one.
 
```

Task 7.2: Strides

```python
# TODO:
# Inspect array strides and explain
A = np.arange(24).reshape(6, 4)
print('shape', A.shape)
print('strides', A.strides) # Strides = It tells how many bytes we need to skip in memory to move to the next element along each dimension.

B  = A[:, ::2]  # slicing every alternate column
print('B shape', B.shape)
print('B strides', B.strides)

# Explanation:
# In a 2D array, the strides indicate how many bytes you need to skip in memory to move to the next row and the next column.
# Here, each element is of type int64, which typically takes up 8 bytes.
# EX:
# (32,8) 
#      0 1 2 3
#      4 5 6 7
# In order to move to the next row (from 0 to 4), we need to skip 4 elements * 8 bytes/element = 32 bytes.
# In order to move to the next column (from 0 to 1), we need to skip 1 element * 8 bytes/element = 8 bytes.
```

Section 8 — Mini Case Study

```python
# TODO:
# Given X (10000, 100):
# - Normalize features
# - Compute covariance
# - Extract top-k eigenvectors

rng = np.random.default_rng(8)
X = rng.standard_normal((10_000, 100))

# Normalize features
mu = X.mean(axis=0, keepdims=True) # mean for each feature , 1 mean per column 
sigma = X.std(axis=0, keepdims=True) # standard deviation for each feature, column-wise standard deviation
Xn = (X - mu) / sigma # normalized features

# Compute covariance
C = (Xn.T @ Xn) / (Xn.shape[0] - 1)  # shape (100, 100)

# Extract top-k eigenvectors 
# Eigen - means how much is the difference in variance along that direction
eigenvalues, eigenvectors = np.linalg.eigh(C)
k = 5  # number of top eigenvectors to extract
top_idx = np.argsort(eigenvalues)[-k:][::-1] # indices of top-k eigenvalues
top_eigenvectors = eigenvectors[:, top_idx]  # shape (100, k)   
print('Top-k eigenvectors shape:', top_eigenvectors.shape)


```
